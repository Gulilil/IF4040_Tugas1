```python

# Installment
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def ScrapDataItem(url):
    html = urlopen(url).read()
    s = BeautifulSoup(html, "html.parser")
    # print(s.prettify())
    return s

def PrintTupleData(tuple):
    for data in tuple:
        print(data, end=" || ")
    print()

def PrintListNewLine(l):
    for i in l:
        print(i)
```

```python
def ScrapAllCompanyList(url):
    tempSoup = ScrapDataItem(url)
    allCategory = tempSoup.findAll("div", {"class" : "box box-edge"})

    hrefArr = []
    for category in allCategory:
      items = category.findAll('div', {'class' : "item"})
      # count = 0
      
      for item in items:
        name = item.find('a').get_text() 
        href = item.find("a")['href']
        # count += 1

        
        tempTuple = (name, href)
        # print(tempTuple)

        hrefArr.append(tempTuple)
      
      # print(count)
    return hrefArr

'''
URL:
https://kpopping.com/profiles/the-companies
'''

companiesList = ScrapAllCompanyList("https://kpopping.com/profiles/the-companies")

PrintListNewLine(companiesList)
```

```python
def ScrapCompanyData(url, name):
    tempSoup = ScrapDataItem(url)
    data = tempSoup.findAll("p", {"class" : "color-blue-green text-uppercase"})

    try:
      hq = data[0].get_text()
    except:
      hq = None

    try:
      country = data[1].get_text()
    except:
      country = None
    
    try:
      foundedSection = tempSoup.find("div", {"class": "col-12"})
      foundedData = foundedSection.findAll("a")
      for i in foundedData:
        try:
          founded = int(i.get_text())
        except:
          continue
    except:
      founded = None
    
    return name, hq, country, founded

'''
Pake list yang ada di atasnya
'''
```

```python
len(companiesList)
```

```python
dataList1 = []
for i in range (300):
  url = 'https://kpopping.com' + companiesList[i][1]
  data = ScrapCompanyData(url, companiesList[i][0])
  dataList1.append(data)
  
  if i % 50 == 0:
    print(i, data)
```

```python
dataList2 = []
for i in range (300, 600):
  url = 'https://kpopping.com' + companiesList[i][1]
  data = ScrapCompanyData(url, companiesList[i][0])
  dataList2.append(data)
  
  if i % 50 == 0:
    print(i, data)
```

```python
dataList3 = []
for i in range (600,900):
  url = 'https://kpopping.com' + companiesList[i][1]
  data = ScrapCompanyData(url, companiesList[i][0])
  dataList3.append(data)
  
  if i % 50 == 0:
    print(i, data)
```

```python
dataList4 = []
for i in range (900, len(companiesList)):
  url = 'https://kpopping.com' + companiesList[i][1]
  data = ScrapCompanyData(url, companiesList[i][0])
  dataList4.append(data)
  
  if i % 50 == 0:
    print(i, data)
```

```python
combineList = dataList1 + dataList2 + dataList3 + dataList4
len(combineList)

numpyArr = np.array(combineList)
df = pd.DataFrame(numpyArr)
df.columns = ['name', 'headquarter', 'country', 'founded_year']
df
```
